```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import urllib3
import time

# Disable SSL warning
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/114.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9"
}

base_url = "https://www.justdial.com/Hyderabad/Hospitals/nct-10253670"

hospital_names = []
contacts = []
ratings = []
addresses = []
urls = []

for page in range(1, 3):  # Scrape 2 pages
    url = f"{base_url}/page-{page}"  # Adjusted URL structure
    response = requests.get(url, headers=headers, verify=False)
    soup = BeautifulSoup(response.text, "html.parser")

    listings = soup.find_all("a", class_="resultbox_title_anchorbox")

    for listing in listings:
        # Hospital Name
        name_tag = listing.find("h3", class_="resultbox_title_anchor")
        name = name_tag.get_text(strip=True) if name_tag else "N/A"

        # URL
        href = listing.get("href", "")
        full_url = "https://www.justdial.com" + href if href else "N/A"

        # Get parent container for rating, contact, etc.
        container = listing.find_parent("li") or listing.find_parent("div")

        # Rating
        rating_tag = container.find("li", class_="resultbox_totalrate") if container else None
        rating = rating_tag.get_text(strip=True) if rating_tag else "N/A"

        # Contact Number
        contact_tag = container.find("span", class_="callcontent") if container else None
        contact = contact_tag.get_text(strip=True) if contact_tag else "N/A"

        # Address
        address_tag = container.find("div", class_="locatcity") if container else None
        address = address_tag.get_text(strip=True) if address_tag else "N/A"

        # Append data
        hospital_names.append(name)
        urls.append(full_url)
        ratings.append(rating)
        contacts.append(contact)
        addresses.append(address)

    time.sleep(2)  # Be polite to the server

# Save to CSV
df = pd.DataFrame({
    "Hospital Name": hospital_names,
    "Contact": contacts,
    "Rating": ratings,
    "Address": addresses,
    "URL": urls
})
df.to_csv("hospitals_justdial.csv", index=False)
print("✅ Scraping Complete — Data saved to 'hospitals_justdial.csv'")
print(df.head())
```
